```python
import nltk
import itertools

def unique_everseen(iterable, key=None):
    seen = set()
    seen_add = seen.add
    if key is None:
        for element in itertools.filterfalse(seen.__contains__, iterable):
            seen_add(element)
            yield element
    else:
        for element in iterable:
            k = key(element)
            if k not in seen:
                seen_add(k)
                yield element
```

```python
trump = open("trump.txt")
trump_corpus = trump.read();

trump_words = nltk.word_tokenize(trump_corpus)
trump_ngrams = nltk.ngrams(trump_words, 1)

trump_unique_everseen = unique_everseen(trump_ngrams)

for word in trump_unique_everseen:
    print(''.join(word))
```

```python
clinton = open("clinton.txt")
clinton_corpus = clinton.read();
clinton_words = nltk.word_tokenize(clinton_corpus)
clinton_ngrams = nltk.ngrams(clinton_words, 1)
clinton_unique_everseen = unique_everseen(clinton_ngrams)


for word in clinton_unique_everseen:
    print(''.join(word))
    
```
